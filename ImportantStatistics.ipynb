# PDF (Probability Density Function):

The PDF represents the probability distribution of a continuous random variable.

The area under the PDF curve over a certain interval represents the probability that the variable falls within that interval.

The PDF does not directly give probabilities; rather, it gives the density of probabilities over the range of possible values.

# CDF (Cumulative Distribution Function):

The CDF gives the cumulative probability that a random variable X is less than or equal to a certain value

The CDF starts at 0 and ends at 1, as it encompasses the entire range of probabilities.
    
The CDF can be obtained by integrating the PDF over the range from negative infinity to x.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data from a normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation
data = np.random.normal(mu, sigma, 1000)

# Plot the histogram of the generated data
plt.hist(data, bins=30, density=True, alpha=0.5, color='b', label='Histogram')

# Plot the PDF (Probability Density Function)
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)  # Calculate the PDF for the normal distribution
plt.plot(x, pdf, 'r', label='PDF')

# Plot the CDF (Cumulative Distribution Function)
cdf = norm.cdf(x, mu, sigma)  # Calculate the CDF for the normal distribution
plt.plot(x, cdf, 'g', label='CDF')

plt.xlabel('Value')
plt.ylabel('Density/Probability')
plt.title('PDF and CDF of Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

# central limit theorem :

The Central Limit Theorem (CLT) is a fundamental concept in statistics that applies to the sampling distribution of the sample mean. It states that if you take repeated samples of a certain size from any population (regardless of the population's distribution) and calculate the mean of each sample, then the distribution of those sample means will be approximately normally distributed, particularly as the sample size increases.

This theorem is incredibly powerful because it allows us to make inferences about population parameters even when we don't know much about the population itself, as long as we have sufficiently large samples. It's the basis for many statistical methods and hypothesis tests, making it a cornerstone of modern statistics.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
population_mean = 100
population_std = 15
sample_size = 1000
num_samples = 1000

# Generate a population with a normal distribution
population = np.random.normal(population_mean, population_std, sample_size)

# Calculate population statistics
pop_mean = np.mean(population)
pop_std = np.std(population)

# Generate samples and calculate means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Calculate statistics for the sample means
sample_means_mean = np.mean(sample_means)
sample_means_std = np.std(sample_means)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')

# Plot the population distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, sample_means_mean, sample_means_std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mean = %.2f,  std = %.2f" % (sample_means_mean, sample_means_std)
plt.title(title)

plt.show()

print("Population Mean:", pop_mean)
print("Population Standard Deviation:", pop_std)
print("Sample Means Mean:", sample_means_mean)
print("Sample Means Standard Deviation:", sample_means_std)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset of delivery times (randomly generated)
np.random.seed(42)  # for reproducibility
delivery_times = np.random.normal(loc=30, scale=10, size=1000)  # mean=30, std=10

# Plot the histogram of delivery times
plt.hist(delivery_times, bins=30, density=True, alpha=0.6, color='b')
plt.title('Distribution of Delivery Times')
plt.xlabel('Delivery Time (minutes)')
plt.ylabel('Probability Density')
plt.show()

# Calculate population statistics
pop_mean = np.mean(delivery_times)
pop_std = np.std(delivery_times)
print("Population Mean:", pop_mean)
print("Population Standard Deviation:", pop_std)

# Now let's take multiple samples from this dataset and calculate the mean of each sample
sample_means = []
num_samples = 1000
sample_size = 50

for _ in range(num_samples):
    sample = np.random.choice(delivery_times, sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')

# Overlay the normal distribution curve with mean and std of sample means
sample_means_mean = np.mean(sample_means)
sample_means_std = np.std(sample_means)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, sample_means_mean, sample_means_std)
plt.plot(x, p, 'k', linewidth=2)

plt.show()

print("Sample Means Mean:", sample_means_mean)
print("Sample Means Standard Deviation:", sample_means_std)
```
